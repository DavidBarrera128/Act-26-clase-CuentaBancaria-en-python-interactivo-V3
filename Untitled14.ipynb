```python
class CuentaBancaria:
    def __init__(self, titular=None, numero_cuenta=None, saldo=0):
        self.titular = titular
        self.numero_cuenta = numero_cuenta
        self.saldo = saldo

    def leerdatos(self, numero_etiqueta):
        """Captura los datos indicando qué número de cuenta se está editando."""
        print(f"\n--- Ingresando datos de la Cuenta Número {numero_etiqueta} ---")
        self.titular = input("Nombre del titular: ")
        self.numero_cuenta = input("Número de cuenta: ")
        # Convertimos a entero para que no aparezcan decimales
        self.saldo = int(float(input("Saldo inicial: ")))

    def mostrardatos(self, numero_etiqueta):
        """Muestra la información con su marcador correspondiente."""
        print(f"\n--- Información de la Cuenta Número {numero_etiqueta} ---")
        print(f"Titular: {self.titular}")
        print(f"Número de Cuenta: {self.numero_cuenta}")
        print(f"Saldo Actual: ${self.saldo}")

# --- Creación de Objetos ---

# Se crea la cuenta 1 y se le pasa el marcador "1"
cuenta1 = CuentaBancaria()
cuenta1.leerdatos(1)

# Se crea la cuenta 2 y se le pasa el marcador "2"
cuenta2 = CuentaBancaria()
cuenta2.leerdatos(2)

# --- Mostrar Resultados ---
print("\n" + "="*40)
print("RESUMEN GENERAL DE CUENTAS")
print("="*40)

cuenta1.mostrardatos(1)
cuenta2.mostrardatos(2)
print("Programa realizado por David Barrera 1388")
```
